### Array math

```python
x = np.array([[1,0,3],[2,1,2]])
y = np.array([[2,1,4],[3,1,1]])

print(x)
print('\n')
print(y)
print('\n')
print(x + y)
print('\n')
print(x - y)
print('\n')
print(x * y)
print('\n')
print(x / y)
print('\n')
print(np.sqrt(x))
print('\n')
print(x.T)
print('\n')
```

```python
print(x)
print('\n')
print(np.sum(x))
print('\n')
print(np.sum(x, axis=0))
print('\n')
print(np.sum(x, axis=1))
print('\n')
```

```python
x = np.array([[1,2],[2,1]])
y = np.array([[0,3],[1,0]])

print(x)
print('\n')
print(y)
print('\n')
print('Perkalian Matrix :')
print(np.dot(x, y))
```

```python
x = np.array([[2,4],[3,1]])
y1 = np.array([[0,3]])
y2 = np.array([[0],[3]])

print(x)
print('\n')
print(y1)
print('\n')
print(y2)
print('\n')
print('Perkalian 1 :')
print(x*y1)

print('Perkalian 2 :')
print(x*y2)
```

## Matplotlib Intro

```python
import matplotlib.pyplot as plt
```

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 10, 0.1)
y = (x-3)**2

# Plot the points using matplotlib
plt.figure(figsize=(15,8))
plt.title('My First Plot')
plt.plot(x, y)
plt.show()
```

```python
# Compute the x and y coordinates for points on sine and cosine curves
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the points using matplotlib
plt.figure(figsize=(15,8))
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('My Second Plot')
plt.legend(['Sine', 'Cosine'])
plt.show()
```

```python
def func(x):
    
    x = x/5
    z = np.sin(10*x)+np.cos(x-3)
    z = z+3
    z = np.log(z)
    
    return z
    
```

```python
# Compute the x and y coordinates for points on sine and cosine curves
x1 = np.arange(0, 40, 0.1)
y_1 = func(x1)

# Plot the points using matplotlib
plt.figure(figsize=(15,8))
plt.plot(x1, y_1)
plt.xlabel('x axis label')
plt.ylabel('y - value')
plt.title('My Third Plot')
plt.legend(['Value Func'])
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve
x = np.array(['satu', 'dua', 'tiga'])
y = np.array([6,7,3])

# Plot the points using matplotlib
plt.figure(figsize=(15,8))
plt.title('My First Plot')
plt.bar(x, y)
plt.show()
```
